```python
import pandas as pd
df = pd.read_csv('student_bmi.csv')
df.head()
df.info()
print(df.columns)
df
```

```python
df['BMI'] = (700*df[' "Weight(Pounds)"']/df[' Height(Inches)"']**2).round(2)
df
print(df.isnull().sum())
df.describe()

```

```python
def bmi_ctg(BMI):
    if BMI<18.5:
        return 'Underweight'
    elif 18.5<=BMI<25:
        return 'Normal'
    elif 25<=BMI<30:
        return 'Overweight'
    else:
        return 'Obese'

df['Category'] = df['BMI'].apply(bmi_ctg)

print(df.columns)

df.rename(columns = {
    ' Height(Inches)"' : 'Height (Inches)',
    ' "Weight(Pounds)"' : 'Weight (Pounds)'
}, inplace=True)
df
df.to_csv('mod_student_bmi.csv', index=False)
df.to_excel('mod_student_bmi.xlsx', index=False)
```
