```python
import pandas as pd
df = pd.read_csv('student_scores.csv')
df
```

```python
#df.sort_values('Math',ascending=False)
```

```python
df['Total'] = df.iloc[:,1:4].sum(axis=1)
df['Average'] = df.iloc[:,1:4].mean(axis=1).round(2)
df
df = df.sort_values('Average',ascending=False)
df.reset_index(drop=True,inplace=True)
df
df.to_csv('modified_student_scores.csv',index=False)
df.to_excel('modified_student_scores.xlsx',index=True)
df

```
